#### Importing the Keras libraries and packages

```

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```
### Creating model and layers
model = Sequential();
# Step 1 - Convolution
model.add(Conv2D(64,(3,3),input_shape=(64,64,3),activation='relu'))
# 64 - no.of filters
# Step 2 - Pooling
model.add(MaxPooling2D(pool_size=(2,2)))
# Step 1 - Convolution
model.add(Conv2D(128,(3,3),activation='relu'))
# Step 2 - Pooling
model.add(MaxPooling2D(pool_size=(2,2)))
# Step 3 - Flattening
model.add(Flatten())
### Hidden and output
# Step 4 - Full connection
model.add(Dense(128, activation='relu'))    ### hidden layer
model.add(Dense(1, activation='sigmoid'))   ## output layer
```

####Compiling the CNN

```

model.compile(optimizer ='adam',loss='binary_crossentropy',metrics=['accuracy'])
```


####Preprocessing of image
#####https://keras.io/api/preprocessing/image/#imagedatagenerator-class

```
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255,
                                  shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
```

```
training_set = train_datagen.flow_from_directory('/content/dataset_new/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

test_set = test_datagen.flow_from_directory('/content/dataset_new/test_set',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

### Train the model

```
model.fit_generator(training_set,
                         steps_per_epoch = 60,
                         epochs = 100,
                         validation_data = test_set,
                         validation_steps = 15)
```

```
from keras.preprocessing import image
import numpy as np
test_img = image.load_img('/content/dataset_new/test_set/cats/cat.4001.jpg',target_size=(64,64)) 
test_img = image.img_to_array(test_img)
test_img = np.expand_dims(test_img,axis=0)
result = model.predict(test_img)
if result[0][0] == '1':
  print ("Dog")
else:
  print ("Cat")
```
